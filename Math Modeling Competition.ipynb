```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

import numpy as np
import pandas as pd
```

```python
race_data = pd.read_csv("./Downloads/Dataset 0-2.csv",  
                  names=["Distance", "Altitude"])
race_data
```

```python
jackson_data = pd.read_csv("./Downloads/Default Dataset-1.csv",
                          names=["x", "y"])
jackson_data
```

```python
X = np.array(race_data["Distance"]).reshape(-1,1)
y = np.array(race_data["Altitude"]).reshape(-1,1)

# Polynomial Regression Model (Degree 11)
for i in range(20):
    polyreg = make_pipeline(PolynomialFeatures(i), LinearRegression())
    polyreg.fit(X,y)
    print("Degree: ", i, " Score: ", polyreg.score(X,y))

    #print(polyreg.named_steps["linearregression"].coef_)
```

```python
polyreg = make_pipeline(PolynomialFeatures(10), LinearRegression())
polyreg.fit(X,y)
print(polyreg.named_steps["linearregression"].coef_)
```
